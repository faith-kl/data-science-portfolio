```python
nlp.select_pipes(enable=["tok2vec", "ner"])

bad_boundaries = {"and", "or", "of", "with", "to", "from", "for", "by", "the", "a", "an"}

ai_noise_pattern = re.compile(
    r'^\s*(ai|AI|LLMs?|GPT-?\d*|SEO|AGI|SGE|Artificial\s*Intelligence)\s*$',
    flags=re.IGNORECASE
)

blocklist = {
    "sms","telecoms","photo","jpeg","covid19","wellness","pms","technology", "business", "latest news", "politics", "contact us",
    "news", "industry news", "events", "archive", "terms & conditions",
    "search", "home", "premium research", "free research", "sections",
    "economy", "sports", "entertainment", "cpu", "gpu", "subscribe",
    "global", "jobs interest", "ftse", "royals uk", "obituaries",
    "daily podcast", "financial solutions", "culture", "puzzles",
    "bookshop", "tickets", "comments", "venture", "time", "back",
    "contest rules", "eeo statement", "public file", "all rights reserved",
    "cancel", "retry", "listen live", "learn more", "sign up", "log in","rights",
    'usa', 'united states', 'uk', 'united kingdom', 'canada', 'australia', 'germany', 'france', 'italy',
    'spain', 'japan', 'china', 'india', 'brazil', 'russia', 'mexico', 'south korea', 'indonesia',
    'turkey', 'netherlands', 'switzerland', 'sweden', 'belgium', 'norway', 'austria', 'poland',
    'argentina', 'saudi arabia', 'thailand', 'egypt', 'iran', 'nigeria', 'south africa', 'colombia',
    'vietnam', 'philippines', 'malaysia', 'pakistan', 'peru', 'romania', 'czech republic', 'chile',
    'greece', 'portugal', 'finland', 'ireland', 'denmark', 'new zealand', 'singapore', 'israel',
    'uae', 'united arab emirates', 'qatar', 'kuwait', 'hong kong', 'taiwan', 'sri lanka', 'bangladesh',
    'kazakhstan', 'ukraine', 'hungary', 'slovakia', 'bulgaria', 'croatia', 'serbia', 'slovenia',
    'lithuania', 'latvia', 'estonia', 'cyprus', 'malta', 'iceland', 'luxembourg', 'albania',
    'bosnia and herzegovina', 'georgia', 'armenia', 'azerbaijan', 'belarus', 'moldova', 'north macedonia',
    'montenegro', 'kosovo', 'afghanistan', 'algeria', 'angola', 'ethiopia', 'ghana', 'kenya',
    'morocco', 'mozambique', 'tanzania', 'tunisia', 'uganda', 'zambia', 'zimbabwe', 'argentina',
    'bolivia', 'ecuador', 'paraguay', 'uruguay', 'venezuela', 'costa rica', 'cuba', 'dominican republic',
    'el salvador', 'guatemala', 'honduras', 'nicaragua', 'panama', 'jamaica', 'trinidad and tobago',
    'bahamas', 'barbados', 'belize', 'guyana', 'haiti', 'suriname', 'korea', 'mexico', 'nepal',
    'oman', 'palestine', 'sudan', 'syria', 'venezuela', 'yemen', 'zambia', 'zimbabwe'
}

ticker_pattern = re.compile(r'^[A-Z]{2,5}$')

def extract_orgs(doc, article_id):
    seen_lower = set()
    final_orgs = []

    for ent in doc.ents:
        if ent.label_ != 'ORG':
            continue

        if len(ent) > 5:
            continue

        if ent[0].text.lower() in bad_boundaries or ent[-1].text.lower() in bad_boundaries:
            continue

        org_clean = ent.text.replace('\n', '').strip().title()

        if len(org_clean) < 3:
            continue

        if ticker_pattern.match(org_clean):
            continue

        if any(blocked in org_clean.lower() for blocked in blocklist):
            continue

        if ai_noise_pattern.match(org_clean):
            continue

        if org_clean.count(',') > 0 or org_clean.count('&') > 1:
            continue

        if org_clean.lower() not in seen_lower:
            seen_lower.add(org_clean.lower())
            final_orgs.append(org_clean)

    return {'article_id': article_id, 'org': final_orgs}
```

```python
from tqdm import tqdm

texts = (text[:3000] for text in df['text_clean'])
ids = df['article_id']

org_entities = []
for doc, article_id in tqdm(
    zip(nlp.pipe(texts, batch_size=128), ids),
    total=len(df)
):
    org_entities.append(extract_orgs(doc, article_id))

    # checkpoint
    if len(org_entities) % 50000 == 0:
        pd.DataFrame(org_entities).to_parquet(
            path + f'ner_checkpoint_{len(org_entities)}.parquet'
        )

org_df = pd.DataFrame(org_entities)
org_df = org_df[org_df['org'].str.len() > 0].copy()
org_df.to_parquet(path + 'ner_orgs_raw.parquet', index=False)
print(f"articles with org: {len(org_df)}")
```
